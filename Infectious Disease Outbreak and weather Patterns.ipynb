```python
import matplotlib.pyplot as plt
import seaborn as sns
years = [2018, 2019, 2020, 2021, 2022]
cases = [120, 105, 131, 117, 90]
plt.figure(figsize=(8, 5))
plt.plot(years, cases, marker='o', color='b')
plt.title("Cases Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Cases")
plt.grid(True)
plt.savefig(r"C:\Users\Melwyn\Pictures\Screenshots\caese over the year.png")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(df['year'], df['Deaths'], color='purple')
plt.title("Deaths Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Death")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='year', y='Temp', data=df, color='', s=100)
plt.title("Scatter Plot ")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df.groupby("Disease")["Deaths"].sum().plot(kind="bar", color="salmon")
plt.xlabel("Disease")
plt.ylabel("Total Deaths")
plt.title("Deaths by Disease Type")
plt.show()
```

```python
plt.figure(figsize=(12,6))
pivot_area = df.pivot_table(values="Cases", index="week_of_outbreak", columns="Disease", aggfunc="sum", fill_value=0)
pivot_area.plot(kind="area", stacked=True, figsize=(12,6), colormap="tab20c", alpha=0.8)
plt.title("Weekly Disease Cases (Stacked Area)", fontsize=14, fontweight='bold')
plt.xlabel("Week of Outbreak")
plt.ylabel("Cases")
plt.xticks(rotation=45)
plt.savefig(r"C:\Users\Melwyn\Pictures\Screenshots\caese over weeks.png")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="Temp", y="Cases", size="preci", hue="Disease", data=df, alpha=0.6, sizes=(50, 500), palette="Set2")
plt.title("Cases vs Temperature (Bubble Size = Precipitation)", fontsize=14, fontweight='bold')
plt.xlabel("Temperature (K)")
plt.ylabel("Cases")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="Disease", y="Temp", data=df, palette="pastel")
plt.title("Temperature Distribution by Disease", fontsize=14, fontweight='bold')
plt.xlabel("Disease")
plt.ylabel("Temperature (K)")
plt.xticks(rotation=45)
plt.show()
```
